```python
import apple
```

```python
c=apple.jit('''
{ mseq ←
    λz.
    {
      add ← λz0.λz1. (z0->1+z1->1, z0->2+z1->2);
      sq ← λ(r,i). (r^2-i^2,2*r*i);
      abs ← λ(r,i). r^2+i^2;
      abs'gen. (0.0,0.0) [add (sq x) z] 64
    };
  [?((>4)@.mseq x)<0,.1.0,.0.0]`{0} ((frange _2.5 1 700) [(x,y)]⊗ (frange _1 1 400))
}
''')
```

```python
arr=c()
```

```python
%timeit c()
```
