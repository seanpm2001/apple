```python
import apple
```

```python
import matplotlib.pyplot as plt
```

```python
# Lemniscate of Bernoulli: https://mathworld.wolfram.com/Lemniscate.html
```

```python
lemni_x=apple.jit("λN. (λt. cos.t%(1+sin.t^2))'(frange (_2*𝜋) (2*𝜋) N)")
```

```python
lemni_y=apple.jit("λN. (λt. sin. t*cos. t%(1+sin. t^2))'(frange (_2*𝜋) (2*𝜋) N)")
```

```python
plt.plot(lemni_x(200),lemni_y(200))
```

```python
# Aztema spiral https://www.2dcurves.com/spiral/spiralat.html
```

```python
aztema_x=apple.jit("λN. (λt. sin.t%t-2*cos.t-t*sin. t)'(frange (1) (1+2*𝜋) N)")
```

```python
aztema_y=apple.jit("λN. (λt. cos.t%t-2*sin.t+t*cos. t)'(frange (1) (1+2*𝜋) N)")
```

```python
plt.plot(aztema_x(200),aztema_y(200))
```
